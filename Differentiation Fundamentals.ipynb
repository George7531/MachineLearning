# process of computing the slope of line segments:
* use formula; m = (y2-y1)/(x2-x1) to calculate slope
* you will get a series of 4 slopes.
* for x axis use (x1,x2) 
* for y axis use (m,m)
* plot the horizontal lines with distinct colors.

```python
# compute the function
n = 5
x = np.linspace(-1,5,n)
y = x**2

fig,ax = plt.subplots(1,2,figsize=(12,4))

# plotting the function
ax[0].plot(x,y,'o-')
ax[0].set_title('Function')
ax[0].set_ylabel('y=f(x)')
ax[0].set_xlabel('x')

# compute the slope of each line:
m = np.zeros(len(x)-1)
for index in range(len(x)-1): #both x and y are of same length
    m[index] = (y[index+1] - y[index])/(x[index+1] - x[index])
    
    
# plotting the slope series as horizontal lines
for index in range(len(m)): # length of m is going to be as same the number of horizontal lines
    ax[1].plot([x[index+1],x[index]],[m[index],m[index]])
    

ax[1].set_title('Series of slopes')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Local Slope(m)')
plt.show()
```

$$ \text{1. compute the global slope using the first and last points. } $$
$$ \text{2. compute the average of the local slopes } $$
$$ \text{3. compare and print both } $$
$$ \text{4. compare } \Delta{y} \text{ with } \frac{\Delta{y} } {\Delta{x}} $$

```python
# 1. compute the global slope using the first and last points.

n = 5
x = np.linspace(-1,5,n)
y = x**2

global_m = (y[-1] - y[0])/(x[-1] - x[0])

local_m = np.zeros(n-1)
only_numerator_of_slopes = np.zeros(n-1)

for index in range(len(local_m)):
    local_m[index] = (y[index+1] - y[index])/(x[index+1] - x[index])
    only_numerator_of_slopes[index] = (y[index+1] - y[index])
    
# average of local slopes
avg_local_m = np.mean(local_m)

# printing global slope and average of local slopes
print('Global slope: {}'.format(global_m))
print('Average of Local Slopes: {}'.format(avg_local_m))


# compare and plot delta y and delta y over delta x
fig,ax = plt.subplots(1,2,figsize=(12,4))

for index in range(len(local_m)):
    ax[0].plot([x[index],x[index+1]],[local_m[index],local_m[index]])
    ax[1].plot([x[index],x[index+1]],[only_numerator_of_slopes[index],only_numerator_of_slopes[index]])
    
ax[0].set_title('$ \\frac{\\Delta{y}}{\\Delta{x}} $',fontsize=30)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y=f(x)')

ax[1].set_title('$ \\Delta{y} $',fontsize=30)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y=f(x)')

plt.show()
```

# formal definintion of slope

# from slope to derivative

$$ m = \frac{y_2 - y_1}{x_2 - x_1} \rightarrow m = \frac{\Delta y}{ \Delta x} $$

$$ \downarrow $$

$$ m = \frac{ f(x + \Delta x) - f(x) } { \Delta x } $$

$$ \downarrow $$

$$ m = \lim_{\Delta x \rightarrow 0} \left[\frac{ f(x + \Delta x) - f(x) } { \Delta x }\right] $$

$$ \downarrow $$

$$ \frac{dy}{dx} = \lim_{\Delta x \rightarrow 0} \left[\frac{ f(x + \Delta x) - f(x) } { \Delta x }\right] $$

$$ \Delta x \text{ are just a series of infinitesimally small points } $$ 


**`derivative == series of local slopes(slopes between infinitisimally small points)`**

# derivative of a constant is zero(0)

$$ \frac{dy}{dx} = \lim_{\Delta x \rightarrow 0} \left[\frac{ f(x + \Delta x) - f(x) } { \Delta x }\right] $$

$$ \downarrow $$

$$ \frac{dy}{dx} = \lim_{\Delta x \rightarrow 0} \left[\frac{c - c } { \Delta x }\right] \rightarrow \frac{0}{\Delta x} $$

$$ \downarrow $$

$$ \frac{dy}{dx} = 0 $$

# various notations of the derivative

# Leibniz derivative notations:

$$ \text{first order: } $$
$$ \frac{dy}{dx} = \frac{df}{dx}  = \frac{d}{dx} f(x) $$
-------------
$$ \text{ second order: } $$
$$ \frac{d^2y}{d^2x} $$

--------------
$$ \text{ chain rule: } $$
$$ \frac{dy}{du} \frac{du}{dx} $$

# difference between secant line and tangent line:

# secant line:
* it passes through two points on a function.
* these points can be distant and far apart.

# tangent line:
* the general talk among mathematicians is that the tangent line touches the function a specific point (a single point).
* but our rigorous approach can prove that that specific point is actually a pair of points that infinitesimally close to each other.
* so it could be more accurate to say it passes through two points on a functions that are infinitesimally close to each other.

# exercise 1: find Derivative

# $$ \frac{5x}{4} + \frac{9}{4} $$

```python
# 1. create a symbolic expression for the function 5x/4 + 9/4
x = sym.symbols('x')
fx = (5*x/4) + 9/4
print(fx)

# 2. plot that line:
p = sym.plot(fx,show=False)
p.show()


# 3.change the location of x=0 line to intersect at y=0:
'''which means substitute 0 for x and subtract fx from the substituted function. this shifts the line down making the 
equation become only fx=5x/4. the 2.5 will be cancelled. so when x=0 the y=0'''
shifted_fx = fx - fx.subs({'x':0})
q = sym.plot(shifted_fx,show=False)
q.show()


# 4. find the derivative for the function:
df_dx = sym.diff(fx,x)
print('The derivative is: {:.4f}'.format(df_dx))
```

# $$ fx = x^2 $$

```python
# 1.create a symbolic expression for fx=x**2
x = sym.symbols('x')
fx = x**2
display(fx)

# 2. plot 
p = sym.plot(fx,show=False) # fx is already passing where x=0 and y=0 so no need of shifting but we'll do it anyway for demo
p.title = 'P'
p.show()

# 3.shift the function to where x=0 and y=0
shifted_fx = fx-fx.subs({'x':0})
q = sym.plot(shifted_fx,show=False)
q.show()

# 4.derivative for the function
df_dx = sym.diff(fx,x)
display(Math(" \\text{ The derivative for the function is: } %s "%(sym.latex(df_dx))))


# 5.plot the derivative with values xx={-1,0,2}
xx_for_df_dx = np.array([-1,0,2])
xx_for_fx = np.linspace(-5,5,100)
numpified_df_dx = sym.lambdify(x,df_dx,'numpy')(xx_for_df_dx)
numpified_fx = sym.lambdify(x,fx,'numpy')(xx_for_fx)
plt.plot(xx_for_fx,numpified_fx)
plt.plot(xx_for_df_dx,numpified_df_dx)
plt.ylim([0,25])
plt.show()


for i in range(len(numpified_df_dx)):
    print('The derivative for dy/dx at x=%s is: %s'%(xx_for_df_dx[i],numpified_df_dx[i]))
```

* when there is no change in function its derivative becomes zero.

# Linearity of Differentiation:

* **Linearity means scaler multiples and scaler sums can be seperated.**



# exercise 1: 

$$ fx = ax^2 + bx^3 + ce^{2x} $$

# goals
$$ \text{1. take the derivative of the whole fx in terms of x } $$
$$ \text{2. break fx into seperate parts: ax^2, bx^3, ce^{2x} and take the derivative of them.} $$
$$ \text{3. prove that step1 == step2 }  $$

```python
a,b,c,x = sym.symbols('a,b,c,x')

fx = a*x**2 + b*x**3 + c*sym.exp(2*x)

#1. take the derivative of the whole fx in terms of x
df_dx = sym.diff(fx,x)
print('The derivative of fx is: ')
display(df_dx)


# 2. break fx into seperate parts: ax^2, bx^3, ce^{2x} and take the derivative of them.

'''please refer exploration, analysis and freethrows section to understand the logic behind the coding and clear 
comprehension!!! '''
parts_of_fx = str(fx).split('+')
c = []

for index in range(len(parts_of_fx)):
    sympified = sym.sympify(parts_of_fx[index])
    diff = sym.diff(sympified,x)
    c.append(diff)
    display(Math('\\text{ Derivative of } %s \\text{ is } %s'%(sym.latex(sympified),sym.latex(diff))))

# prove step1 == step2
display(Math('\\text{The combination of individual components is: } %s'%(sym.latex(c[0]+c[1]+c[2]))))
```

# Exercise 2: scaler multiples of Derivatives:

$$ \text{prove } $$
$$ \frac{d}{dx} (ax^2) = a \left(\frac{d}{dx}x^2 \right)$$

```python
a,x = sym.symbols('a,x')

# taking derivatives with respect to x
lhs = sym.diff(a*x**2,x)
rhs = a*sym.diff(x**2)

# printing the results
display(Math('\\text{Lhs = }%s'%sym.latex(lhs)))
display(Math('\\text{Rhs = }%s'%sym.latex(rhs)))
```

# conclusion:

henced proved operations of the derivatives are Linear(allows for both scaler addition and scaler multiplication to be seperated)

# exercise 3: Full Linearity Test

$$ \text{prove } $$
$$ \frac{d}{dx} (ax^2 + b\cos{x}) = a \left(\frac{d}{dx}x^2 \right) + b \left(\frac{d}{dx}\cos{x} \right)  $$

```python
a,b,x = sym.symbols('a,b,x')

lhs = sym.diff(a*x**2 + b*sym.cos(x),x)
rhs = a*sym.diff(x**2,x) + b* sym.diff(sym.cos(x),x)

# printing the results
display(Math('\\text{Lhs = }%s'%sym.latex(lhs)))
display(Math('\\text{Rhs = }%s'%sym.latex(rhs)))
```

# Conclusion:

* **Hence proved that the Linearity of the derivatives hold True that the addition(shifiting) and multiplication(scaling) of the components of the derivative of the function can be seperated and amalgamated together**
