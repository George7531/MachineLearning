# Theoritical normal distribution vs empirical or experimental normal distribution

```python
import numpy as np
import sympy as sym
from IPython.display import display,Math
sym.init_printing()
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
x = np.linspace(-4,4,10001)

# theoritical normal distribution
theonormdist = stats.norm.pdf(x)



# empirical normal distribution
n = 400000 # number of samples

sample_data = np.zeros(n)

for i in range(n):
    sample_data[i] = np.random.randn()


# plotting

plt.hist(sample_data,density=True) # density converts the plot from showing counts to probability
plt.plot(x,theonormdist,'r',linewidth=3)
plt.xlabel('Data Values')
plt.ylabel('Probability')


plt.show()

```
